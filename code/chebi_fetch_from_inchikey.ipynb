# Fetch CHEBI from InChiKey

## Overall Toxicity Fetch Chebi

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import rdkit as Chem

# Load the Excel file
df = pd.read_excel('../data/Final S2 v3.xlsx', sheet_name='Overall Toxicity Curated Data', engine='openpyxl')

# Print column names to verify
print("Column names:", df.columns)

# Add a new column for the CHEBI ID
df['CHEBI_ID'] = ''

#testing

# Set up the Selenium WebDriver
driver = webdriver.Chrome()  # You may need to specify the path if it's not in your PATH
url = 'https://www.ebi.ac.uk/chebi/init.do'

# Iterate through all rows
for index, row in df.iterrows():
    inchikey = row['InChIKey']  # Ensure this matches the actual column name in your Excel file
    driver.get(url)
    
    try:
        # Wait for the search box to be present and send the InChIKey
        search_box = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, '//*[@id="content"]/div[2]/form/input[1]'))
        )
        search_box.send_keys(inchikey)
        search_box.send_keys(Keys.RETURN)
        
        # Wait for the results page to load and display the CHEBI ID
        chebi_id_element = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, '/html/body/div[1]/div[1]/table[2]/tbody/tr[1]/td/table[1]/tbody/tr/td/table/tbody/tr[1]/td[2]/table/tbody/tr[3]/td[2]/b'))
        )
        chebi_id = chebi_id_element.text.strip()
        df.at[index, 'CHEBI_ID'] = chebi_id
        
    except Exception as e:
        print(f"Error for InChIKey {inchikey}: {e}")
        df.at[index, 'CHEBI_ID'] = 'Not Found'
        time.sleep(2)  # Add a small delay before the next iteration if there's an error

# Save the updated DataFrame to a new Excel file
df.to_excel('../data/overall_trimester_toxicity_with_chebi.xlsx', index=False)  # Ensure the file extension is .xlsx

driver.quit()
```
